```python
import numpy as np

Z1 = 10 + 5j
Z2 = 8 - 3j
Z3 = 6 + 2j

Z_series = Z1 + Z2 + Z3

Z_parallel = 1 / (1/Z1 + 1/Z2 + 1/Z3)

print("Impedance Z1 =", Z1)
print("Impedance Z2 =", Z2)
print("Impedance Z3 =", Z3)

print("Total Series Impedance =", Z_series)
print("Total Parallel Impedance =", Z_parallel)
```
