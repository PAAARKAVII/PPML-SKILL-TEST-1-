```python
import pandas as pd

# EEG sample data
data = {
    "Electrode1": [10, 12, 11, 13, 15],
    "Electrode2": [8, 9, 10, 9, 11],
    "Electrode3": [14, 13, 15, 16, 14]
}

df = pd.DataFrame(data)

print("EEG Data:\n", df)

print("\nMean Values:\n", df.mean())
print("\nVariance:\n", df.var())
print("\nPeak Values:\n", df.max())
```
