```python
import numpy as np

# Step data for 10 days
steps = np.array([5000, 6200, 7000, 8000, 9000, 6500, 7200, 8100, 9500, 4000])

print("Total steps:", steps.sum())
print("Average steps:", steps.mean())
print("Maximum steps:", steps.max())

print("\nHigh activity days(>8000):")
for i in range(len(steps)):
    if steps[i] > 8000:
        print("Day", i + 1, ":", steps[i])
```
