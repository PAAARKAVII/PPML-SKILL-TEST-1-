```python
import numpy as np

scores = np.array([45, 78, 56, 89, 90, 67, 34, 23, 88, 76, 54, 61, 72, 48, 59, 92, 81, 39, 64, 70])

average = np.mean(scores)
highest = np.max(scores)
lowest = np.min(scores)

print("Average Score:", average)
print("Lowest Score:", highest)  # Note: Label in image says Lowest but variable is highest
print("Lowest Score:", lowest)

pass_mark = 50

print("\nPass/Fail Status")
for i, score in enumerate(scores):
    if score >= pass_mark:
        print(f"Student {i+1}: Pass")
    else:
        print(f"Student {i+1}: Fail")

print("\nGrades")
for i, s in enumerate(scores):
    if s >= 90:
        grade = "A"
    elif s >= 75:
        grade = "B"
    elif s >= 60:
        grade = "C"
    elif s >= 50:
        grade = "D"
    else:
        grade = "F"
    print("Student", i+1, "Grade:", grade)
```
